```python
# Part 1
with open("data/2025_02.txt") as f:
    input_str = f.read().strip()

input = [[int(n) for n in s.split("-")] for s in input_str.split(",")]

out = 0

for [start, end] in input:
    for n in range(start, end + 1):
        n_str = str(n)
        if len(n_str) % 2 == 1:
            continue

        pivot = len(n_str) // 2
        for i in range(pivot):
            if n_str[i] != n_str[pivot + i]:
                break
        else:
            out += n

print(out)
```

```python
# Part 2
with open("data/2025_02.txt") as f:
    input_str = f.read().strip()

input = [[int(n) for n in s.split("-")] for s in input_str.split(",")]
out = 0

for [start, end] in input:
    for n in range(start, end + 1):
        n_str = str(n)
        max_len = len(n_str) // 2

        for i in range(max_len):
            # is there a repeating sequence of this length?
            seq = n_str[0 : i + 1]
            len_seq = len(seq)

            if len(n_str) % len_seq != 0:
                continue

            next_start = len_seq
            next_end = next_start + len_seq

            # does this sequence repeat all the way to the end?
            while next_end <= len(n_str):
                if n_str[next_start:next_end] != seq:
                    break

                next_start = next_end
                next_end = next_start + len_seq
            else:
                out += n
                break

print(out)
```
