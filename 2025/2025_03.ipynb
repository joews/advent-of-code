```python
# Part 1
with open("data/2025_03.txt") as f:
    input_str = f.read().strip()

input = input_str.strip().splitlines()
out = 0

for row in input:
    max_1 = "0"
    max_2 = "0"

    for i, c in enumerate(row):
        if c > max_1 and i < len(row) - 1:
            max_1 = c
            max_2 = "0"
        elif c > max_2:
            max_2 = c

    out += int(max_1) * 10 + int(max_2)

print(out)
```

```python
# Part 2
with open("data/2025_03.txt") as f:
    input_str = f.read().strip()

input = input_str.strip().splitlines()

N = 12
out = 0

for row in input:
    selection = ["0"] * N

    for i, c in enumerate(row):
        # what's the highest place this digit can go?
        remaining_in_row = len(row) - i
        max_place = max(0, N - remaining_in_row)

        # put this digit in the highest position it can displace
        for j in range(max_place, N):
            if c > selection[j]:
                selection[j] = c
                selection[j + 1 :] = ["0"] * (N - j - 1)
                break

    out += int("".join(selection))

print(out)
```
