```python
# Part 1
with open("data/2025_03.txt") as f:
    input_str = f.read().strip()

input = input_str.strip().splitlines()
out = 0

for row in input:
    max_1 = "0"
    max_2 = "0"

    for i, c in enumerate(row):
        if c > max_1 and i < len(row) - 1:
            max_1 = c
            max_2 = "0"
        elif c > max_2:
            max_2 = c

    out += int(max_1) * 10 + int(max_2)

print(out)
```
