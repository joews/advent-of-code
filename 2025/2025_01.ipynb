```python
# Part 2
with open("data/2025_01.txt") as f:
    input_str = f.read()

lines = input_str.strip().split("\n")
input = [[line[0] == "L" and -1 or 1, int(line[1:])] for line in lines]

current = 50
out = 0

for dir, delta in input:
    next = current + dir * delta
    for c in range(current, next, dir):
        if c % 100 == 0:
            out += 1

    current = next % 100

print(out)
```
