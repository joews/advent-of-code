```python
# Part 1
from math import dist, inf

with open("data/2025_09.txt") as f:
    input_str = f.read().strip()

input = [
    tuple([int(c) for c in line.split(",")]) for line in input_str.strip().splitlines()
]


def area(a, b):
    return (abs(a[0] - b[0]) + 1) * (abs(a[1] - b[1]) + 1)


max_area = 0
for i, a in enumerate(input):
    for j, b in enumerate(input):
        if a != b:
            ab = area(a, b)
            if ab > max_area:
                max_area = ab


print(max_area)
```
