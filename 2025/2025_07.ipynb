```python
# Part 1
with open("data/2025_07.txt") as f:
    input_str = f.read().strip()

rows = [list(row) for row in input_str.strip().splitlines()]
rays = set([rows[0].index("S")])

out = 0
for row in rows:
    splits = [i for i, c in enumerate(row) if c == "^"]

    if splits:
        current_rays = list(rays)

        for ray in current_rays:
            if ray in splits:
                out += 1
                rays.remove(ray)

            for split in splits:
                rays.add(split - 1)
                rays.add(split + 1)

print(out)
```

```python
# Part 2
from collections import Counter

with open("data/2025_07.txt") as f:
    input_str = f.read().strip()

rows = [list(row) for row in input_str.strip().splitlines()]
LAST_ROW_INDEX = len(rows) - 1

rays = Counter({rows[0].index("S"): 1})

for row_num, row in enumerate(rows):
    splits = [i for i, c in enumerate(row) if c == "^"]

    for split in splits:
        if split in rays:
            n = rays[split]
            del rays[split]

            rays[split - 1] += n
            rays[split + 1] += n

print(rays.total())
```
