```python
from functools import reduce
import operator

with open("data/2025_06.txt") as f:
    input_str = f.read().strip()

input = [row.split() for row in input_str.strip().splitlines()]
out = 0

for col in range(len(input[0])):
    func = input[-1][col] == "+" and operator.add or operator.mul
    args = [int(row[col]) for row in input[:-1]]

    out += reduce(func, args)

print(out)
```
