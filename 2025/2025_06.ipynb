```python
from functools import reduce
from operator import add, mul

OPS = {"+": add, "*": mul}

with open("data/2025_06.txt") as f:
    input_str = f.read().strip()

input = [row.split() for row in input_str.strip().splitlines()]
out = 0

for col in range(len(input[0])):
    op = input[-1][col]
    args = [int(row[col]) for row in input[:-1]]

    out += reduce(OPS[op], args)

print(out)
```

```python
# Part 2
from functools import reduce
from operator import add, mul

OPS = {"+": add, "*": mul}

with open("data/2025_06.txt") as f:
    input_str = f.read().strip()

input_lines = input_str.splitlines()

# use the last row to find the column breaks
last_line = input_lines[-1]
splits = [i - 1 for i, c in enumerate(last_line) if i > 0 and c != " "]
col_breaks = [-1, *splits, len(last_line) + 1]
col_ranges = list(zip([i + 1 for i in col_breaks[:-1]], col_breaks[1:]))

# Build a list of justified strings per column
rows = [[line[r[0] : r[1]] for r in col_ranges] for line in input_lines]
cols = [list(row) for row in zip(*rows)]

# Do the math
total = 0
for col in cols:
    max_place = len(col[0]) - 1
    op = col[-1].strip()
    args = []

    for place in range(max_place, -1, -1):
        arg_list = [row[place] for row in col[:-1]]
        args.append(int("".join(arg_list)))

    result = reduce(OPS[op], args)
    total += result

print(total)
```
