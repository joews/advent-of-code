```python
# Part 2
with open("data/2025_05.txt") as f:
    input_str = f.read().strip()

input = input_str.strip().split("\n\n")
input_ranges = [[int(r) for r in row.split("-")] for row in input[0].splitlines()]


# Assume `ranges` is sorted
# Collapse adjacent overlapping pairs in one pass to avoid mid-iteration mutation
# Nasty! out of time to make it less gnarly.
def collapse(sorted_ranges):
    i = 0
    next_ranges = []

    while i < len(sorted_ranges):
        r = sorted_ranges[i]
        next_ranges.append(r)

        if len(sorted_ranges) == i + 1:
            break

        rn = sorted_ranges[i + 1]
        if rn[0] >= r[0] and rn[0] <= r[1]:
            if rn[1] > r[1]:
                r[1] = rn[1]
            i += 2
        else:
            i += 1

    return next_ranges


ranges = sorted(input_ranges)

# Keep collapsing until all overlapping pairs are merged
for _ in range(1000):
    last_ranges = ranges
    ranges = collapse(ranges)
    if len(ranges) == len(last_ranges):
        break
else:
    print("hit limit")

total = 0
for r in ranges:
    range_total = r[1] - r[0] + 1
    total += range_total

print(total)
```
