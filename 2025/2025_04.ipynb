```python
# Part 1
with open("data/2025_04.txt") as f:
    input_str = f.read().strip()

grid = [list(row) for row in input_str.strip().splitlines()]
max_y = len(grid)
max_x = len(grid[0])

out = 0


def adj(y_in, x_in):
    result = 0
    for j in [-1, 0, 1]:
        for i in [-1, 0, 1]:
            if j == 0 and i == 0:
                continue

            yt = y_in + j
            xt = x_in + i

            if 0 <= yt < max_y and 0 <= xt < max_x and grid[yt][xt] == "@":
                result += 1

    return result


for y, row in enumerate(grid):
    for x, c in enumerate(row):
        if grid[y][x] == "@" and adj(y, x) < 4:
            out += 1

print(out)
```
