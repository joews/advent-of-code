```python
# Part 1
lines = input_str.strip().splitlines()

obstacles = set()
gy, gx = -1, -1
g = None

for (y, row) in enumerate(lines):
    for (x, c) in enumerate(row):
        if c == "#":
            obstacles.add((y, x))
        elif c == "^" or c == ">":
            gy, gx = y, x
            g = c

visited = set()

while gy < len(lines) and gy >= 0 and gx < len(lines[0]) and gx >= 0:
    visited.add((gy, gx))

    if g == "^":
        if (gy - 1, gx) in obstacles:
            g = ">"
        else:
            gy -= 1
    elif g == ">":
        if (gy, gx + 1) in obstacles:
            g = "v"
        else:
            gx += 1 
    elif g == "v":
        if (gy + 1, gx) in obstacles:
            g = "<"
        else:
            gy += 1 
    elif g == "<":
        if (gy, gx - 1) in obstacles:
            g = "^"
        else:
            gx -= 1 

print(len(visited))


```

```python
# Debug
def dump():
    rows = []
    for y in range(len(lines)):
        cells = []
        rows.append(cells)
        for x in range(len(lines[0])):
            if y == gy and x == gx:
                cells.append(g)
            elif (y, x) in obstacles:
                cells.append("#")
            else:
                cells.append(".")
                         
    print("\n".join(["".join(r) for r in rows]))
```
