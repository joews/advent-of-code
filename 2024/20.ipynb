```python
# Part 1
start = (-1, -1)
end = (-1, 1)

grid = [list(line) for line in input_str.strip().splitlines()]
for y, row in enumerate(grid):
    for x, c in enumerate(row):
        if c == "S":
            start = (y, x)
        if c == "E":
            end = (y, x)

# (count of moves to this cell, dy to next cell, dx to next cell)
cell_data = {}

(y, x) = start
moves = 0
closed = set()

# Get baseline path length and set traversal direction for each cell
while grid[y][x] != "E" and moves < 100000:
    closed.add((y, x))

    for dy, dx in [(1, 0), (0, 1), (-1, 0), (0, -1)]:
        ny, nx = y + dy, x + dx
        if grid[ny][nx] != "#" and (ny, nx) not in closed:
            # ouch
            if (y, x) == start:
                cell_data[start] = (0, dy, dx)

            (y, x) = (ny, nx)
            moves += 1
            cell_data[y, x] = (moves, dy, dx)
            break

result = 0
for y, row in enumerate(grid):
    for x, c in enumerate(row):
        if c != "#" or x == 0 or y == 0 or x == len(row) - 1 or y == len(grid) - 1:
            continue

        for dy, dx in [(0, 1), (1, 0)]:
            if grid[y - dy][x - dx] in [".", "E", "S"] and grid[y + dy][x + dx] in [
                ".",
                "E",
                "S",
            ]:
                n1 = cell_data[(y - dy, x - dx)]
                n2 = cell_data[(y + dy, x + dx)]
                s = abs(n1[0] - n2[0]) - 2
                if s >= 100:
                    result += 1
                continue

print(result)
```

```python
# Part 2
start = (-1, -1)
grid = [list(line) for line in input_str.strip().splitlines()]
for y, row in enumerate(grid):
    for x, c in enumerate(row):
        if c == "S":
            start = (y, x)

(y, x) = start
path = [start]

while grid[y][x] != "E":
    closed.add((y, x))

    for dy, dx in [(1, 0), (0, 1), (-1, 0), (0, -1)]:
        ny, nx = y + dy, x + dx
        if grid[ny][nx] != "#" and (ny, nx) not in path:
            (y, x) = (ny, nx)
            path.append((y ,x))
            break

MIN_SAVING = len(grid) > 20 and 100 or 50
MAX_CHEAT = 20
total = 0

for (moves1, (y1, x1)) in enumerate(path):
    for moves2 in range(moves1 + 0, len(path)):
        (y2, x2) = path[moves2]
        cheat_distance = abs(y2 - y1) + abs(x2 - x1)
        if cheat_distance > MAX_CHEAT:
            continue

        move_diff = moves2 - moves1 - cheat_distance
        if move_diff >= MIN_SAVING:
            total += 1

print(total)
```
