```python
# Part 1
grid = []

for line in input_str.strip().splitlines():
    row = []
    grid.append(row)
    for c in line:
        row.append(c)

h = len(grid)
w = len(grid[0])


def adjacent(cell):
    y, x = cell
    deltas = [(1, 0), (0, 1), (-1, 0), (0, -1)]

    return [
        (y + dy, x + dx) for dy, dx in deltas if 0 <= y + dy < h and 0 <= x + dx < w
    ]


def get_region(start_y, start_x):
    c = grid[start_y][start_x]
    perimeter = 0
    region_cells = set()
    stack = [(start_y, start_x)]

    while stack:
        (y, x) = stack.pop()
        if (y, x) in region_cells:
            continue
        region_cells.add((y, x))

        if y == 0 or y == h - 1:
            perimeter += 1
        if x == 0 or x == w - 1:
            perimeter += 1

        for ay, ax in adjacent((y, x)):
            if grid[ay][ax] != c:
                perimeter += 1
            else:
                stack.append((ay, ax))

    return (perimeter, region_cells)


seen = set()
result = 0

for y, row in enumerate(grid):
    for x, c in enumerate(row):
        if (y, x) in seen:
            continue

        (perimeter, cells) = get_region(y, x)
        result += perimeter * len(cells)
        seen.update(cells)

print(result)
```
