```python
# Part 2
from collections import defaultdict

antennas = defaultdict(list)
antinodes = set()

lines = input_str.strip().splitlines()
for y, line in enumerate(lines):
    for x, c in enumerate(line):
        if c != ".":
            antennas[c].append((y, x))

h = len(lines)
w = len(lines[0])


def add(a, b):
    return (a[0] + b[0], a[1] + b[1])


def sub(a, b):
    return (a[0] - b[0], a[1] - b[1])


for key, sites in antennas.items():
    if len(sites) > 1:
        antinodes.update(sites)

    for i, src in enumerate(sites):
        for dst in sites[i + 1 :]:
            dy = dst[0] - src[0]
            dx = dst[1] - src[1]

            last_src = src
            last_dst = dst
            for i in range(100):
                a = sub(last_src, (dy, dx))
                b = add(last_dst, (dy, dx))

                for n in [a, b]:
                    if 0 <= n[0] < h and 0 <= n[1] < w:
                        antinodes.add(n)

                last_src = a
                last_dst = b

print(len(antinodes))
```
