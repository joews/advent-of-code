```python
# Setup
def parse(input_str):
    rules_str, pages_str = input_str.split("\n\n")

    rules = [ints(line, "|") for line in rules_str.strip().splitlines()]
    pages = [ints(line, ",") for line in pages_str.strip().splitlines()]

    return rules, pages

def ints(str_of_ints, delim):
    return [int(s) for s in str_of_ints.split(delim)]


    
```

```python
from z3 import *


def p(page_num):
    return f"p{page_num}"

class Vars:
    state = {}

    def get(self, p_num):
        key = p(p_num)
        if (key not in self.state):
            self.state[key] = Int(key)

        return self.state[key]
    
    def __repr__(self):
        return str(self.state.keys())

# Part 1
def part1():
    rules, pages = parse(input_str)
    vars = Vars()

    result = 0
    for page_seq in pages:
        # print(page_seq)

        # Many times faster to create a new solve with the rules constraints
        # vs creating once and cloning
        solver = Solver()
        for (p1, p2) in rules:
            # The order constraints are consistent with a given page_seq
            # but inconsistent across page_seqs. I can see where part 2 is going...
            if (p1 in page_seq and p2 in page_seq):
                v1 = vars.get(p1)
                v2 = vars.get(p2)

                # print(v1, "<", v2)
                solver.add(v1 < v2)

        for i in range(1, len(page_seq)):
            v1 = vars.get(page_seq[i - 1])
            v2 = vars.get(page_seq[i])

            # print("seq", v1, "<", v2)
            solver.add(v1 < v2)

        solve_result = solver.check()
        # print("\t", solver.check())
        if (solve_result == sat):
            mid = page_seq[len(page_seq) // 2]
            result += mid

    print(result)

part1()
```
