```python
# Part 1
lines = input_str.strip().splitlines()
chunks = lines[0].split(", ")
designs = lines[2:]


def get_prefixes(str):
    prefixes = [c for c in chunks if str.startswith(c)]
    return prefixes


def find_design(design):
    stack = []
    stack.append((0, design))

    while stack:
        (c, target) = stack.pop()

        for p in get_prefixes(target):
            if c + len(p) == len(design):
                return True

            stack.append((c + len(p), target[len(p) :]))

    return False


result = len([d for d in designs if find_design(d)])
print(result)
```
