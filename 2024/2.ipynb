```typescript
function isPairSafe(a, b, direction) {
  const diff = Math.abs(a - b);

  if (diff < 1 || diff > 3) {
    return false;
  }

  if (direction === 1 && a - b < 0) {
    return false;
  }

  if (direction === -1 && a - b > 0) {
    return false;
  }

  return true;
}

function isRowSafe(row) {
  if (row[0] === row[1]) {
    return false;
  }

  const direction = row[0] < row[1] ? 1 : -1;
  return row.every((_, i) =>
    i == 0 || isPairSafe(row[i], row[i - 1], direction)
  );
}

function part1() {
  const parsedInput = input
    .trim()
    .split("\n")
    .map((r) => r.trim().split(" ").map((e) => parseInt(e, 10)));

  const result1 = parsedInput.filter(isRowSafe).length;
  console.log(result1);
}


part1();
```
