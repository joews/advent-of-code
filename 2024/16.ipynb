```python
# Part 1
import heapq

walls = set()
goal = (-1, -1)
start = (-1, -1)
orientation = (0, 1)

for (grid_y, row) in enumerate(input_str.splitlines()):
    for (grid_x, c) in enumerate(row):
        if c == "#":
            walls.add((grid_y, grid_x))
        elif c == "S":
            start = (grid_y, grid_x)
        elif c == "E":
            goal = (grid_y, grid_x)

def adjacent(node):
    y, x = node
    oy, ox = orientation
    out = []

    for (dy, dx) in [(1, 0), (0, 1), (-1, 0), (0, -1)]:
        if ox == 0 and oy == -1 * dy or oy == 0 and ox == -1 * dx:
            continue

        ny, nx = y + dy, x + dx
        if (ny, nx) not in walls:
            cost = (dy, dx) == orientation and 1 or 1001
            out.append(((ny, nx), cost))

    return out

def distance(a, b):
    return abs(a[1] - b[1]) + abs(a[0] - b[0])

open = []
closed = set()
parents = {}
node_cost = { start: 0 }

heapq.heappush(open, (0, start))

i = 0
while len(open):# and i < 1000:
    i +=1 
    f, node = heapq.heappop(open)
    closed.add(node)

    # print("enter", node, open)

    # parent = parents[node]
    if node in parents:
        orientation = (node[0] - parents[node][0], node[1] - parents[node][1])

    if node == goal:
        print(node_cost[goal])
        break

    for (n, cost_n) in adjacent(node):
        if (n in closed):
            continue

        node_cost[n] = node_cost[node] + cost_n
        parents[n] = node

        f_score = node_cost[n] + distance(n, goal)
        heapq.heappush(open, (f_score, n))

print("end")



    
```
