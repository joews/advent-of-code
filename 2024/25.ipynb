```python
# Part 1
blocks = input_str.strip().split("\n\n")
keys = set()
locks = set()


def parse_block(lines):
    counts = [0] * 5
    for l in lines:
        for i, c in enumerate(l):
            if c == "#":
                counts[i] += 1

    return tuple(counts)


for block in blocks:
    lines = block.split("\n")
    if lines[0] == "#####":
        keys.add(parse_block(lines[1:]))
    else:
        locks.add(parse_block(reversed(lines[:-1])))


def check(key, lock):
    for i in range(5):
        if key[i] + lock[i] > 5:
            return False
    return True


result = 0
for key in keys:
    for lock in locks:
        if check(key, lock):
            result += 1


print(result)
```
