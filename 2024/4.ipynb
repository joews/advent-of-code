```typescript
// Part 1
const TARGET = "XMAS";

const TARGET_FORWARD_PATTERN = new RegExp(TARGET, "g");
const TARGET_REVERSE_PATTERN = new RegExp(
    TARGET.split("").reverse().join(""),
    "g",
);

function transpose(matrix) {
    return matrix[0].map((_, colIndex) => matrix.map((row) => row[colIndex]));
}

function reflectY(matrix) {
    return matrix.map((row) => [...row].reverse());
}

function rowStrings(matrix) {
    return matrix.map((m) => m.join(""));
}

// TODO is the other way the LTR of the transpose?!
function diagonalStrings(matrix) {
    const output = [];

    const h = matrix.length;
    const w = matrix[0].length;

    // See what diagonal we can make from each start square
    for (let y = 0; y < h; y++) {
        // Cover the top-right half of the puzzle with diagonals from the first row.
        // Cover the bottom-left half of the puzzle with diagonals from the first column.
        const stop = y == 0 ? w : 1;

        for (let x = 0; x < stop; x++) {
            const diagonal = [];

            let j = y;
            let i = x;

            while (j < h && i < w) {
                diagonal.push(matrix[j][i]);

                i++;
                j++;
            }

            if (diagonal.length >= TARGET.length) {
                output.push(diagonal.join(""));
            }
        }
    }

    return output;
}

function countMatches(haystack: string) {
    let count = 0;
    const forwardMatches = haystack.match(TARGET_FORWARD_PATTERN);
    const reverseMatches = haystack.match(TARGET_REVERSE_PATTERN);

    if (forwardMatches) {
        count += forwardMatches.length;
    }

    if (reverseMatches) {
        count += reverseMatches.length;
    }

    return count;
}

function sumMatches(haystacks: string[]) {
    return haystacks.reduce((sum, h) => sum + countMatches(h), 0);
}

function part1() {
    // Precompute all of the rows, columns and diagonals then match the target
    // strong forwards and in reverse.
    const matrix = input.trim().split("\n").map((row) => row.trim().split(""));

    const hStrs = rowStrings(matrix);
    const vStrs = rowStrings(transpose(matrix));
    const diagonalLtr = diagonalStrings(matrix);
    const diagonalRtl = diagonalStrings(reflectY(matrix));

    const result = sumMatches(hStrs) + sumMatches(vStrs) +
        sumMatches(diagonalLtr) + sumMatches(diagonalRtl);

    console.log(result);
}

part1();
```

```typescript
// Part 2
function isHit(m, y, x) {
    if (m[y][x] != "A") {
        return false;
    }

    let ltr = (m[y - 1][x - 1] === "M" && m[y + 1][x + 1] === "S") ||
        (m[y - 1][x - 1] === "S" && m[y + 1][x + 1] === "M");

    let rtl = (m[y - 1][x + 1] === "M" && m[y + 1][x - 1] === "S") ||
        (m[y - 1][x + 1] === "S" && m[y + 1][x - 1] === "M");

    return ltr && rtl;
}

function part2() {
    const matrix = input.trim().split("\n").map((row) => row.trim().split(""));
    const h = matrix.length;
    const w = matrix[0].length;

    let result = 0;

    for (let y = 1; y < h - 1; y++) {
        for (let x = 1; x < w - 1; x++) {
            if (isHit(matrix, y, x)) {
                result += 1;
            }
        }
    }

    console.log(result);
}

part2();
```
