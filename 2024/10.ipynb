```python
# Part 1
grid = []

for y, line in enumerate(input_str.strip().splitlines()):
    grid.append([int(c) for c in line])


def adjacent(y, x):
    deltas = [(1, 0), (0, 1), (-1, 0), (0, -1)]
    height, width = len(grid), len(grid[0])

    return [
        (y + dy, x + dx)
        for dy, dx in deltas
        if 0 <= y + dy < height
        and 0 <= x + dx < width
        and grid[y + dy][x + dx] == grid[y][x] + 1
    ]


def count_trails(y, x):
    stack = [(y, x)]
    seen = set()
    result = 0

    while stack:
        (y, x) = stack.pop()
        if (y, x) in seen:
            continue
        seen.add((y, x))

        if grid[y][x] == 9:
            result += 1
            continue

        stack.extend(adjacent(y, x))

    return result


result = sum(
    count_trails(y, x)
    for y, row in enumerate(grid)
    for x, c in enumerate(row)
    if c == 0
)

print(result)
```
