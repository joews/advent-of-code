```python
# Part 1
from collections import defaultdict

adjacent = defaultdict(set)

for row in input_str.strip().splitlines():
    (src, dst) = row.split("-")
    adjacent[src].add(dst)
    adjacent[dst].add(src)

threes = set()

for src, hop1 in adjacent.items():
    for n1 in hop1:
        for n2 in hop1:
            if n1 >= n2:
                continue

            if n2 in adjacent[n1] and (src[0] == "t" or n1[0] == "t" or n2[0] == "t"):
                three = tuple(sorted([src, n1, n2]))
                threes.add(three)


print(len(threes))
```

```python
# Part 2
from collections import defaultdict

adjacent = defaultdict(set)

for row in input_str.strip().splitlines():
    (src, dst) = row.split("-")
    adjacent[src].add(dst)
    adjacent[dst].add(src)


def is_clique(nodes):
    for n in nodes:
        for o in nodes:
            if n != o and o not in adjacent[n]:
                return False
    return True


def grow_clique(clique, candidates):
    for c in candidates:
        next_clique = clique | {c}
        if is_clique(next_clique):
            result = grow_clique(next_clique, candidates - {c})
            if result:
                return result

    # couldn't add any more nodes
    return clique


biggest_clique = set()
for k in adjacent.keys():
    clique = grow_clique({k}, adjacent[k])
    if len(clique) > len(biggest_clique):
        biggest_clique = clique


print(",".join(sorted(list(biggest_clique))))
```
