```python
# Part 1
from collections import defaultdict
import re

def parse(input_str):
    seeds_str, *map_sections = input_str.split("\n\n")
    input_seeds = ints(seeds_str.split(": ")[1])

    nodes_by_type = defaultdict(list)
    next_types = {}

    for section in map_sections:
        (map_nodes, src_type, dst_type) = parse_map(section)
        for n in map_nodes:
            nodes_by_type[n["type"]].append(n)

        next_types[src_type] = dst_type

    return (input_seeds, nodes_by_type, next_types)


def parse_map(map_str):
    types, *lines = map_str.split("\n")
    (src_type, dst_type) = re.findall(r"(\w+)-to-(\w+)", types)[0]
    edge_ranges = [ints(line) for line in lines]

    section_nodes = []
    for edge_range in edge_ranges:
        dst_start, src_start, length = edge_range

        section_nodes.append(node(src_type, src_start, length, dst_start))

    return (section_nodes, src_type, dst_type)


def node(type, min, length, dst_min):
    return {"type": type, "min": min, "length": length, "dst_min": dst_min}

def ints(str_of_ints):
    return [int(x) for x in str_of_ints.split(" ")]

def find_node(nodes, type, id):
    # Is there a node with an explicit mapping?
    for n in nodes[type]:
        if id >= n["min"] and id < (n["min"] + n["length"]):
            return n

    # If not, use an implicit mapping
    return node(type, id, 1, id)


def traverse(nodes, type, id):
    # print("traverse", type, id)

    # Have we reached the end?
    if type == "location":
        return id
    else:
        node = find_node(nodes, type, id)
        offset = id - node['min']

        next_node_id = node['dst_min'] + offset

        return traverse(nodes, next_types[type], next_node_id)

(input_seeds, nodes, next_types) = parse(input)
locations = [traverse(nodes, "seed", seed_id) for seed_id in input_seeds]
print("Part 1 result:", min(locations))

```
