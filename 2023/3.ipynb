```python
# Part 1

matrix = [[c for c in line] for line in input.splitlines()]

# State
part_ids = []

current_digits = []
is_part = False
row = 0
col = 0

def is_symbol(c):
    return not c.isdigit() and c != '.'

def accept_part():
    if len(current_digits) > 0:
        part_ids.append(int(''.join(current_digits)))
        current_digits.clear()
        # print("Found part", part_ids[-1])

def check_for_part(row, col):
    # If we have just finished a run of digits, check if they are a part.
    # If we have no digits, nothing to do.
    if len(current_digits) == 0:
        return

    # The dot terminates a sequence of digits. Check if it's a part.
    # Check all of the x cells which are within the bounds of the matrix.
    #
    # x x x x x
    # x 1 2 3 . <- we are here
    # x x x x x
    #
    start_col = max(col - len(current_digits) - 1, 0)
    end_col = min(col, len(matrix[row]) - 1)

    # Check the previous row, unless we're on the first row
    if (row > 0):
        for t in range(start_col, end_col + 1):
            if is_symbol(matrix[row - 1][t]):
                accept_part()
                return

    # Check the next row, unless we're on the last row
    if (row < len(matrix) - 1):
        for t in range(start_col, end_col + 1):
            if is_symbol(matrix[row + 1][t]):
                accept_part()
                return

    # Check for a symbol before the current digits
    if col > 0 and is_symbol(matrix[row][start_col]):
        accept_part()
        return

    # If we made it this far, the digits are not a part.
    # print("Not a part", "".join(current_digits))
    current_digits.clear()


def on_digit(c):
    # print("[digit]", c)
    current_digits.append(c)

def on_dot(c, row, col):
    # print("[dot]", c)
    check_for_part(row, col)

def on_symbol(c):
    # print("[symbol]", c)
    # if there were digits to the left of this symbol, they are a part id
    if (len(current_digits) > 0):
        accept_part()

def on_row_end():
    # print("[row end]")
    check_for_part(row, col + 1)
    # print("-------------")
    


for row in range(len(matrix)):
    current_digits = []
    is_part = False

    for col in range(len(matrix[row])):
        c = matrix[row][col]

        if c.isdigit():
            on_digit(c)
        elif c == '.':
            on_dot(c, row, col)
        else:
            on_symbol(c)

    on_row_end()

print(sum(part_ids))

        
```
