```python
# Part 1
def parse(input_str):
    return [list(line) for line in input_str.splitlines()]

# Mutates grid
def expand_grid(grid):
    # expand rows
    y = 0
    while y < len(grid):
        if all([c == "." for c in grid[y]]):
            clone = grid[y][:]
            grid.insert(y, clone)
            y += 1
        y += 1
        
    # expand columns
    x = 0
    while x < len(grid[0]):
        col = get_col(x)
        if all([c == "." for c in col]):
            for y in range(len(grid)):
                grid[y].insert(x, ".")

            x += 1

        x += 1
    
def get_col(x):
    col = []
    for y in range(len(grid)):
        col.append(grid[y][x])

    return col   

def set_col(x, value):
    col = []
    for y in range(len(grid)):
        grid[y][x] = value

def dump(grid):
    for y in range(len(grid)):
        for x in range(len(grid[0])):
            print(grid[y][x], end="")
        print("")
    

def distance(a, b):
    return abs(b[0] - a[0]) + abs(b[1] - a[1])

# Go
grid = parse(input_str)
expand_grid(grid)
# dump(grid)

cells = []
for y in range(len(grid)):
    for x in range(len(grid[0])):
        if grid[y][x] == "#":
            cells.append((y, x))

sum = 0
for i in range(len(cells)):
    for j in range(i + 1, len(cells)):
        sum += distance(cells[i], cells[j])

print("Part 1: ", sum)
```
