```python
# Part 1
ranks = ["2", "3", "4", "5", "6", "7", "8", "9", "T", "J", "Q", "K", "A"]


def parse(input):
    hands = []
    for line in input.splitlines():
        (cards, bid) = line.split(" ")
        hands.append((cards, [rank(c) for c in cards], int(bid)))

    return hands


def rank(c):
    return ranks.index(c)


def key(hand):
    ranks = hand[1]
    return [rank_hand(ranks), ranks]


def rank_hand(hand):
    runs = get_runs(hand)
    try:
        return [[1, 1, 1, 2], [1, 2, 2], [1, 1, 3], [2, 3], [1, 4], [5]].index(runs) + 1
    except:
        return 0


# Return an array containing the repeats of character sets in a list.
# Always returns the shortest repeats first.
# [1, 2, 1, 2] => [2, 2]
# [2, 2, 3, 2] => [2, 3]
# [6, 6, 6, 6] => [4]
# [3, 5, 2, 9] => [1, 1, 1, 1]
def get_runs(hand):
    s = sorted(hand)
    runs = []
    start = 0
    for i in range(1, len(hand)):
        if s[i] != s[i - 1]:
            runs.append(i - start)
            start = i

    runs.append(len(s) - start)
    return sorted(runs)


def go():
    input = parse(input_str)
    ranked = sorted(input, key=key)

    sum = 0
    for i, (x, y, bid) in enumerate(ranked):
        sum += (i + 1) * bid

    return sum


print("Part 1:", go())
```

```python
# Part 2
# Monkey patch away!
ranks = ["J", "2", "3", "4", "5", "6", "7", "8", "9", "T", "Q", "K", "A"]


# Count 0 (J) as a wildcard
def get_runs(hand):
    # Remove zeros for now. Later they will be included in the longest run.
    s = [e for e in sorted(hand) if e != 0]
    num_zeros = len(hand) - len(s)

    runs = []
    start = 0

    for i in range(1, len(s)):
        if s[i] != s[i - 1]:
            runs.append(i - start)
            start = i

    runs.append(len(s) - start)
    sorted_runs = sorted(runs)

    # Include zeros in the longest run
    sorted_runs[-1] += num_zeros
    return sorted_runs

print("Part 2:", go())
```
