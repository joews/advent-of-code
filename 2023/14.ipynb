```python
# Part 2
def move_n(grid, x, y):
    row = y - 1
    while row >= 0 and grid[row][x] != "#":
        if grid[row][x] == ".":
            grid[row][x] = "O"
            grid[row + 1][x] = "."
        row -= 1


def move_s(grid, x, y):
    row = y + 1
    while row < len(grid) and grid[row][x] != "#":
        if grid[row][x] == ".":
            grid[row][x] = "O"
            grid[row - 1][x] = "."
        row += 1


def move_w(grid, x, y):
    col = x - 1

    while col >= 0 and grid[y][col] != "#":
        if grid[y][col] == ".":
            grid[y][col] = "O"
            grid[y][col + 1] = "."
        col -= 1


def move_e(grid, x, y):
    col = x + 1

    while col < len(grid[y]) and grid[y][col] != "#":
        if grid[y][col] == ".":
            grid[y][col] = "O"
            grid[y][col - 1] = "."
        col += 1


def tilt_n(grid):
    for y in range(1, len(grid)):
        for x in range(len(grid[y])):
            if grid[y][x] == "O":
                move_n(grid, x, y)


def tilt_s(grid):
    for y in range(len(grid) - 2, -1, -1):
        for x in range(len(grid[y])):
            if grid[y][x] == "O":
                move_s(grid, x, y)


def tilt_w(grid):
    for y in range(len(grid)):
        for x in range(1, len(grid[y])):
            if grid[y][x] == "O":
                move_w(grid, x, y)


def tilt_e(grid):
    for y in range(len(grid)):
        for x in range(len(grid[y]) - 2, -1, -1):
            if grid[y][x] == "O":
                move_e(grid, x, y)


def load_n(grid):
    result = 0
    for y in range(0, len(grid)):
        load = len(grid) - y
        result += sum([1 for r in grid[y] if r == "O"]) * load

    return result


# Manual step:
# Run a few cycles and eyeball for a repeat pattern
grid = parse(input_str)
loads = []

# Run to spot a cycle visually
for i in range(200):
    tilt_n(grid)
    tilt_w(grid)
    tilt_s(grid)
    tilt_e(grid)

    load = load_n(grid)
    loads.append(load)

    # print(i, "->", load)

# Find the solution
CYCLES = 1000000000

# From inspection:
cycle_start = 95
cycle_len = 22

# TODO: work out where the -1 fits in
solution_step = (CYCLES - cycle_start - 1) % cycle_len

print("Part 2:", loads[cycle_start + solution_step])
# 98887 low
# 98894 yes
```
