```python
from math import floor

# Part 1
def parse(input_str):
    grids = input_str.split("\n\n")
    return [parse_grid(g) for g in grids]

def parse_grid(grid):
    return [list(line) for line in grid.splitlines()]

def dump_grid(grid, visited=set()):
    for y in range(len(grid)):
        for x in range(len(grid[y])):
            print(grid[y][x], end="")
        print("")

def dump_transposed_grid(grid, visited=set()):
    for x in range(len(grid[0])):
        c = col(grid, x)
        for y in range(len(c)):
            print(c[y], end="")
        print("")

def col(grid, x):
    return [grid[y][x] for y in range(len(grid))]

def find_v_line(grid):
    width = len(grid[0])
    # print("widrh", width)

    for x in range(1, width):
        if col(grid, x - 1) == col(grid, x):
            # print(f"found pair {x-1},{x}")
            n = 0
            while x - 1 - n >= 0 and x + n < width:
                # print(f"Now checking::: {x-1-n},{x+n}")
                if col(grid, x - 1 - n) != col(grid, x + n):
                    # print("break")
                    break
                n = n + 1
                # print("check next::: ", x - 1 -n, x + n)
            else:
                return x
    return -1

def find_h_line(grid):
    height = len(grid)

    for y in range(1, height):
        # print(f"Checking {y}")
        if grid[y - 1] == grid[y]:
            print(f"found pair {y-1},{y}")
            n = 1
            while y - 1 - n >= 0 and y + n < height:
                print(f"Now checking {y-n},{y+n}")
                if grid[y - 1 - n] != grid[y + n]:
                    break
                n = n + 1
            else:
                return y
    return -1

h_only = 0
v_only = 0
both = 0
neither = 0

def find_line(grid):
    global h_only
    global v_only
    global both
    global neither

    v = find_v_line(grid)
    h = find_h_line(grid)

    if v > 0 and h > 0:
        both += 1
        print("---------------------")
        print("Both lines found in grid")
        dump_grid(grid)
        print("---------------------")  
        return 0

    if v > 0:
        v_only += 1
        return v
    elif h > 0:
        h_only += 1
        return h * 100
    else:
        neither += 1
        print("---------------------")
        print("No line found in grid")
        dump_grid(grid)
        print("---")  
        dump_transposed_grid(grid)
        print("---------------------")  
        return 0

# grid = parse_grid("""
# ..####.
# #..#.##
# .###.#.
# #.##..#
# ###...#
# .#.#...
# .#.#...
# ###...#
# #.##..#
# .###.#.
# #..####
# ..####.
# ..####.
# """.strip())

grids = parse(input_str)
grid = grids[1]

# print("h", find_h_line(grid))
# print("v", find_v_line(grid))

sum = 0
for grid in grids:
    # print(find_line(grid))
    sum += find_line(grid)

print("h", h_only)
print("v", v_only)
print("both", both)
print("neither", neither)
print("total", h_only + v_only + both + neither)
print("grids", len(grids))
print("Part 1:", sum)

```
