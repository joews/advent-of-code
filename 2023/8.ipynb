```python
# Part 1
import re


def parse(input_str):
    lines = input_str.splitlines()
    turns = list(lines[0])

    nodes = {}

    for line in lines[2:]:
        match = re.match("([A-Z]+) = \(([A-Z]+), ([A-Z]+)\)", line)
        (id, l, r) = match.groups()
        nodes[id] = (id, l, r)

    return (turns, nodes)



def part1():
    (turns, nodes) = parse(input_str)
    n_steps = len(turns)

    next_node_id = "AAA"
    step = 0

    while next_node_id != "ZZZ":
        n = nodes[next_node_id]
        branch = turns[step % n_steps] == "L" and 1 or 2
        next_node_id = n[branch]

        step += 1

    return step

print(part1())
```
